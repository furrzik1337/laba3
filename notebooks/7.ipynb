```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Создание сетки координат:
x = np.linspace(-5, 5, 50) # Создаем 50 точек от -5 до 5 по оси x
y = np.linspace(-5, 5, 50) # Создаем 50 точек от -5 до 5 по оси y
X, Y = np.meshgrid(x, y) # Создаем сетку (матрицы X и Y, где каждая ячейка соответствует координате (x,y))

# 2. Вычисление значений Z:
Z = np.sin(np.sqrt(X**2 + Y**2)) # Вычисляем z = sin(√(x² + y²))

# 3. Создание 3D-графика:
fig = plt.figure(figsize=(10, 8))  # Создаем окно для графика
ax = fig.add_subplot(111, projection='3d') # Создаем 3D-подграфик

# 4. Построение поверхности:
surf = ax.plot_surface(X, Y, Z, cmap='viridis')  # Рисуем поверхность

# Другие варианты графиков:
# wireframe = ax.plot_wireframe(X,Y,Z)  # Построение каркасной модели
# surface = ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap = cm.viridis) #Более гибкий способ отрисовки

# 5. Настройка графика:
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('График функции z = sin(√(x² + y²))')
fig.colorbar(surf)  # Добавляем цветовую шкалу

# 6. Отображение графика:
plt.show()
```
