```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1.   :
x = np.linspace(-5, 5, 50) #  50   -5  5   x
y = np.linspace(-5, 5, 50) #  50   -5  5   y
X, Y = np.meshgrid(x, y) #   ( X  Y,      (x,y))

# 2.   Z:
Z = np.sin(np.sqrt(X**2 + Y**2)) #  z = sin((x + y))

# 3.  3D-:
fig = plt.figure(figsize=(10, 8))  #    
ax = fig.add_subplot(111, projection='3d') #  3D-

# 4.  :
surf = ax.plot_surface(X, Y, Z, cmap='viridis')  #  

#   :
# wireframe = ax.plot_wireframe(X,Y,Z)  #   
# surface = ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap = cm.viridis) #   

# 5.  :
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('  z = sin((x + y))')
fig.colorbar(surf)  #   

# 6.  :
plt.show()
```
