```python
import matplotlib.pyplot as plt

# 1. :
grades = ['', '', '', '']
counts = [20, 35, 30, 15]

# 2.   :
plt.figure(figsize=(8, 6))

plt.pie(counts, labels=grades, autopct='%1.1f%%', startangle=90,
        explode=(0, 0.1, 0, 0),  # ""  ""  
        shadow=True,            #  
        colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']) #   

# :
#   - counts:  
#   - labels:  
#   - autopct:     (, '%1.1f%%' -    )
#   - startangle: ,      (90 - ""  )
#   - explode: ,   ""  (,  )
#   - shadow:     
#   - colors:     .

# 3.  :
plt.title('  ')

# 4.  :
plt.show()
```
