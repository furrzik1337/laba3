```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Генерация случайных данных:
np.random.seed(42)  # Для воспроизводимости результатов
x = np.random.rand(50) * 20 - 10  # 50 случайных x-координат в диапазоне [-10, 10]
y = np.random.rand(50) * 20 - 10  # 50 случайных y-координат в диапазоне [-10, 10]

# 2. Задание цветов и размеров:
colors = x      # Цвет зависит от x-координаты
sizes = y * 10 # Размер зависит от y-координаты (умножаем, чтобы лучше видеть разницу)

# Важно: размеры точек должны быть положительными
sizes = np.abs(sizes) + 1  # Добавляем 1, чтобы все размеры были положительными и не были слишком маленькими

# 3. Построение диаграммы рассеяния:
plt.figure(figsize=(8, 6)) # Опционально: задаем размер графика
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis') # c - цвет, s - размер, alpha - прозрачность, cmap - цветовая схема

# 4. Настройка графика:
plt.xlabel('x')
plt.ylabel('y')
plt.title('Диаграмма рассеяния с зависимостью цвета и размера от координат')
plt.colorbar(label='Координата x (цвет)') # Добавляем цветовую шкалу для понимания соответствия цвета x

# 5. Отображение графика:
plt.show()

```
