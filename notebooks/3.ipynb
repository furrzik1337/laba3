```python
import numpy as np
import matplotlib.pyplot as plt

# 1.   :
np.random.seed(42)  #   
x = np.random.rand(50) * 20 - 10  # 50  x-   [-10, 10]
y = np.random.rand(50) * 20 - 10  # 50  y-   [-10, 10]

# 2.    :
colors = x      #    x-
sizes = y * 10 #    y- (,    )

# :     
sizes = np.abs(sizes) + 1  #  1,          

# 3.   :
plt.figure(figsize=(8, 6)) # :   
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis') # c - , s - , alpha - , cmap -  

# 4.  :
plt.xlabel('x')
plt.ylabel('y')
plt.title('        ')
plt.colorbar(label=' x ()') #        x

# 5.  :
plt.show()

```
